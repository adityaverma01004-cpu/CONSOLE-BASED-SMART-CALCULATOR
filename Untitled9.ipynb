```python
import numpy as np
a = int(input("enter the number of students:"))
stu_list = []
for i in range(a):
  b = int(input("enter the marks of student:"))
  stu_list.append(b)
  if b>=90 and b<=100:
    print("you get EX")
  elif b>=80 and b<=90:
    print("you get A")
  elif b>=70 and b<=80:
    print("you get B")
  elif b>=60 and b<=70:
    print("you get C")
  else:
    print("FAIL")
stu_array = np.array(stu_list)

print(f"the minimum score is {min(stu_list)}")
print(f"the maximum score is {max(stu_list)}")
print(f"the average score is {np.mean(stu_list)}")
print(f"the total score is {sum(stu_list)}")
```
